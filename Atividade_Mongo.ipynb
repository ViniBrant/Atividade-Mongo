```python
import pymongo
from pymongo import MongoClient

pymongo.version
```

```python
client = pymongo.MongoClient("mongodb+srv://admin:admin@cluster0.zdshd.mongodb.net/anatel?retryWrites=true&w=majority")
```

```python
db.minas.count_documents({}) # Quantos documentos possui a coleção minas?
```

```python
list(db.minas.find().limit(5)) # Mostre os 5 primeiros documentos da coleção.
```

```python
db.minas.find_one(sort=[("QtdeSolic", -1)]) #Selecione o documento que possui o maior número de solicitações QtdeSolic na Anatel.
```

```python
valueMin = db.minas.find_one(sort=[("QtdeSolic", 1)]) # Qual é o menor número de solicitações QtdeSolic na Anatel?
valueMin['QtdeSolic']
```

```python
minValue = db.minas.find_one(sort=[("QtdeSolic", 1)])
value = minValue['QtdeSolic']
list(db.minas.find({ "QtdeSolic": value })) # Mostre todos os documentos com o menor número de solicitações QtdeSolic na Anatel.
```

```python
reclamacoes = list(db.minas.find({ "Tipo": "Reclamação"})) #Mostre todos os tipos de documentos com o atributo igual a reclamação
len(reclamacoes)
```

```python
list(db.minas.distinct('Tipo')) #Quais diferentes valores encontramos em Tipo?
```
